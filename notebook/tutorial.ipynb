# Create a workbook

```python
import os
from pathlib import Path
from openpyxl import Workbook, load_workbook
from openpyxl.utils import get_column_letter
```

```python
current_path = Path(os.getcwd())
parent_path = current_path.parent
out_path = parent_path / 'out'
current_path, parent_path, out_path
```

```python
wb = Workbook()
ws1 = wb.active
ws1.title = 'Range Names'
for row in range(1, 40):
    ws1.append(range(60))
ws2 = wb.create_sheet(title='Pi')
ws2['F5'] = 3.14
ws3 = wb.create_sheet(title='Data')
for row in range(10, 20):
    for col in range(27, 54):
        _ = ws3.cell(column=col, row=row, value=f'{get_column_letter(col)}')
wb.save(out_path / 'tmp1.xlsx')
```

# Formula

读取 excel 文件，该文件中包含公式，其中：D8 单元格公式为 `=sum(D3:D7)`，尝试修改 D7 单元格，测试 D8 单元格的数值是否会跟着变化

```python
f = parent_path / 'data' / 'formula.xlsx'
wb = load_workbook(f)
ws = wb.active
ws['D7'] = 100
wb.save(f)
```

通过上面的测试，修改了 D7 单元格的值，D8 单元格（是一个公式）的数值会发生变化，但 openpyxl 包并不会计算公式的值。

下面的代码罗列出所支持的公式

```python
from openpyxl.utils import FORMULAE

for fml in FORMULAE:
    print(f'{fml}')
```

# Flod（outline）

```python
f = parent_path / 'data' / 'flod.xlsx'
wb = Workbook()
ws = wb.active
ws.column_dimensions.group('A', 'D', hidden=True)
ws.row_dimensions.group(1, 10, hidden=True)
wb.save(f)
```
